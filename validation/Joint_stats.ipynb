## PDF

```python
d = np.loadtxt("diff23_pdf.dat")
nupdf = d[:, 0]
pdfd = d[:, 1]
pdfd_e = d[:, 2]
d1 = np.loadtxt("pm23_pdf2.dat")
pdf_dark = d1[:, 1]
```

```python
r1 = np.load('final_pdf_71.npy',allow_pickle=True)
r2 = np.load('final_pdf_72.npy',allow_pickle=True)
r3 = np.load('final_pdf_73.npy',allow_pickle=True)
r4 = np.load('final_pdf_74.npy',allow_pickle=True)
r5 = np.load('final_pdf_75.npy',allow_pickle=True)
r6 = np.load('final_pdf_76.npy',allow_pickle=True)
```

```python
#np.arange(-0.24875, 0.99875+0.0025,0.0025)
x1 = r1[0][0]
x2 = r2[0][0]
x3 = r3[0][0]
x4 = r4[0][0]
x5 = r5[0][0]
x6 = r6[0][0]
x=list(x1)+list(x2)+list(x3)+list(x4)+list(x5)+list(x6)
```

```python
pdfb1 = r1[:,1]
pdfb2 = r2[:,1]
pdfb3 = r3[:,1]
pdfb4 = r4[:,1]
pdfb5 = r5[:,1]
pdfb6 = r6[:,1]

pdfd1 = r1[:,2]
pdfd2 = r2[:,2]
pdfd3 = r3[:,2]
pdfd4 = r4[:,2]
pdfd5 = r5[:,2]
pdfd6 = r6[:,2]

diff1 = (pdfb1-pdfd1).mean(axis=0)
diff2 = (pdfb2-pdfd2).mean(axis=0)
diff3 = (pdfb3-pdfd3).mean(axis=0)
diff4 = (pdfb4-pdfd4).mean(axis=0)
diff5 = (pdfb5-pdfd5).mean(axis=0)
diff6 = (pdfb6-pdfd6).mean(axis=0)

std1 = (pdfb1-pdfd1).std(axis=0)
std2 = (pdfb2-pdfd2).std(axis=0)
std3 = (pdfb3-pdfd3).std(axis=0)
std4 = (pdfb4-pdfd4).std(axis=0)
std5 = (pdfb5-pdfd5).std(axis=0)
std6 = (pdfb6-pdfd6).std(axis=0)

pdf_bar = np.array(list(diff1)+list(diff2)+list(diff3)+list(diff4)+list(diff5)+list(diff6))
pdf_dmo = np.array(list(pdfd1.mean(axis=0))+list(pdfd2.mean(axis=0))+list(pdfd3.mean(axis=0))+list(pdfd4.mean(axis=0))+list(pdfd5.mean(axis=0))+list(pdfd6.mean(axis=0)))
```

```python
std_ratio= np.array(list(std1)+list(std2)+list(std3)+list(std4)+list(std5)+list(std6))
```

```python
x1_pdf=x
```

```python
mean_ratio = (pdf_bar)/pdf_dmo
#std1 = (pdf_bar-pdf_dmo).std(axis=0)
#std2 = pdf_dmo.std(axis=0)

#std_peak = np.sqrt((std1/(pdf_bar-pdf_dmo).mean(axis=0))**2+(std2/pdf_dmo.std(axis=0))**2)
#std_ratio = np.abs(mean_ratio)*std_peak

plot(x1_pdf,mean_ratio,lw=3, c="red",label="our result")
#plot(nupdf, pdfd/pdf_dark, lw=3, c="C0", label="Paper")

fill_between(x1_pdf, mean_ratio + std_ratio/pdf_dmo, mean_ratio, alpha=0.4, color='lightsteelblue')
fill_between(x1_pdf, mean_ratio- std_ratio/pdf_dmo, mean_ratio, alpha=0.4, color='lightsteelblue')

#fill_between(nupdf, (pdfd-pdfd_e)/pdf_dark, (pdfd+pdfd_e)/pdf_dark,color="green", alpha=0.2)

xlabel(r'$\kappa$',size=14)
ylabel(r'$\Delta P/P$',size=14)
xlim(-0.03, 0.1)
yscale('symlog', linthreshy=0.1)
ylim(-1,1)
yticks([-1, -0.1, 0, 0.1, 1])
legend()
grid(True)
show()
```
